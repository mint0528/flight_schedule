```python
# the path between flight1 and flight2
def ff_connect(flight_id1,flight_id2):
    value1 = fs_0505[fs_0505.flight_id == flight_id1]
    value2 = fs_0505[fs_0505.flight_id == flight_id2]
    
    if value1['de_ap'].values == value2['la_ap'].values and value1['de_time'].values > value2['de_time'].values:
        return 1
    return 0

print(ff_connect(1241, 132))
print(ff_connect(1241, 2070))
print(ff_connect(1241,519))
```

```python
indexes = fs_0505['flight_id'].values
```

```python
import numpy as np
```

```python
ffp = pd.DataFrame(data=np.zeros((511, 511)), index=indexes, columns=indexes) # flight to flight pxxxxx matrix
```

```python
for id1 in indexes:
    for id2 in indexes:
        ffp.loc[id1, id2] = ff_connect(id1, id2)
```
